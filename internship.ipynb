```python
import pandas as pd
import numpy as np

data = {
    'Product': ['Laptop', 'Mouse', np.nan, 'Keyboard', 'Laptop'],
    'Revenue': [50000, 2000, 10000, np.nan, 50000]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

# Fill missing values
df['Product'] = df['Product'].fillna('Unknown')
df['Revenue'] = df['Revenue'].fillna(df['Revenue'].mean())

# Remove duplicates
df = df.drop_duplicates()

print("\nCleaned Dataset:")
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Mouse'],
    'Revenue': [50000, 2000, 5000, 60000, 3000]
}

df = pd.DataFrame(data)

top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)

print(top_products)

plt.bar(top_products.index, top_products.values, color='skyblue')
plt.title('Top Selling Products')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['2025-01-10', '2025-02-15', '2025-03-20',
             '2025-04-12', '2025-05-18', '2025-06-25'],
    'Revenue': [50000, 60000, 55000, 70000, 90000, 85000]
}

df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.strftime('%B')

monthly_sales = df.groupby('Month')['Revenue'].sum()

print(monthly_sales)

plt.plot(monthly_sales.index, monthly_sales.values,
         marker='o', linewidth=3)

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data
products = ['Laptop', 'Mouse', 'Keyboard', 'Headphones']
revenue = [110000, 5000, 7000, 15000]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [50000, 60000, 55000, 70000, 90000, 85000]

plt.figure(figsize=(12,5))

# Bar Chart
plt.subplot(1,2,1)
plt.bar(products, revenue, color='orange')
plt.title("Top Selling Products")
plt.xlabel("Products")
plt.ylabel("Revenue")

# Line Chart
plt.subplot(1,2,2)
plt.plot(months, sales, marker='o', color='green')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)

plt.tight_layout()
plt.show()
```
